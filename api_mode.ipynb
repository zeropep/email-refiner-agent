```python
import requests

BASE_URL = "http://127.0.0.1:8000"
APP_NAME = "travel_advisor_agent"
USER_ID = "u_123"
SESSION_ID = "ed70ff03-b3f5-4bef-a5a9-c5cbd2861d05"
```

```python
response = requests.post(f"{BASE_URL}/apps/{APP_NAME}/users/{USER_ID}/sessions")

print(response.json())
```

```python
message = {
    "appName": APP_NAME,
    "userId": USER_ID,
    "sessionId": SESSION_ID,
    "newMessage": {
        "parts": [
            {"text": "I also want to know the currency exchange rate"}
        ],
        "role": "user"
    }
}

response = requests.post(f"{BASE_URL}/run", json=message)

# print(response.json())
```

```python
data = response.json()

for event in data:
    content = event.get("content")
    parts = content.get("parts")
    for part in parts:
        function_call = part.get("functionCall", None)
        if function_call:
            print(function_call.get("name"))
        text = part.get("text", None)
        if text:
            print(text)
    print("+" * 50)
```
